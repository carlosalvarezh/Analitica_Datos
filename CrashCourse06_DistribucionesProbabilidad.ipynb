## Introducción

En el vasto campo de la estadística y la probabilidad, las [distribuciones de probabilidad](https://en.wikipedia.org/wiki/Probability_distribution) juegan un papel fundamental como modelos matemáticos que nos permiten comprender y analizar cómo se comportan las [variables aleatorias](https://en.wikipedia.org/wiki/Random_variable). Las distribuciones son herramientas esenciales que ayudan a los estadísticos, científicos de datos y analistas a dar sentido a los datos, tomar decisiones informadas y hacer predicciones sobre fenómenos naturales y sociales. En este contexto, una distribución se refiere a la manera en que se distribuyen o dispersan los valores de una variable aleatoria en un conjunto de datos.

## Definición

Una distribución en estadística y probabilidad es una descripción sistemática y matemática de cómo se agrupan o esparcen los valores de una variable aleatoria dentro de un conjunto de datos. Representa las probabilidades de ocurrencia de diferentes valores o resultados posibles que puede tomar dicha variable. En esencia, una distribución proporciona información sobre la tendencia central, la dispersión y la forma en que se distribuyen los datos.

## Usos

Las distribuciones tienen una amplia variedad de aplicaciones en estadística y probabilidad:

- ***Descripción de Datos:*** Permiten resumir y describir la variabilidad en los datos observados, lo que es esencial para comprender el comportamiento de una variable.
<p>&nbsp;</p>

- ***Modelado de Fenómenos Naturales y Humanos:*** Modelar y entender cómo se comportan variables en el mundo real. Por ejemplo, la distribución normal se utiliza para describir la altura de las personas.
<p>&nbsp;</p>

- ***Inferencia Estadística:*** Ayudan en la toma de decisiones basadas en datos a través de pruebas de hipótesis y estimaciones de intervalos de confianza. Las distribuciones como la [$t$ de Student](https://en.wikipedia.org/wiki/Student%27s_t-distribution) y la distribución [chi-cuadrado](https://en.wikipedia.org/wiki/Chi-squared_distribution) son esenciales en estas pruebas.
<p>&nbsp;</p>

- ***Predicción y Simulación:*** Se utilizan para prever resultados futuros en base a datos históricos. Esto es crucial en campos como la economía y la meteorología.
<p>&nbsp;</p>

- ***Control de Calidad:*** En la industria y la manufactura, se emplean distribuciones para evaluar la calidad y la variabilidad de los productos.
<p>&nbsp;</p>

- ***Investigación Científica:*** En la investigación científica, ayudan a analizar datos experimentales y determinar si los resultados son estadísticamente significativos.
<p>&nbsp;</p>

- ***Análisis de Riesgo:*** Se aplican en la evaluación y gestión de riesgos financieros, ambientales y de seguridad.
<p>&nbsp;</p>

- ***Procesamiento de Señales:*** En campos como la ingeniería de comunicaciones y el procesamiento de imágenes, se utilizan distribuciones para analizar señales y ruido.
<p>&nbsp;</p>

- ***Machine Learning:*** Son esenciales en algoritmos de aprendizaje automático, donde se utilizan para modelar y predecir datos.

## Clasificación

Las distribuciones pueden clasificarse en dos categorías principales en función de la naturaleza de las variables aleatorias que representan:

###  Definiciones Básicas

Veamos primero unos conceptos importantes:

- ***Variable aleatoria, $X$:*** Una variable aleatoria es una función que asigna valores numéricos a los resultados posibles de un experimento aleatorio. Es una representación matemática de los resultados de un evento incierto. Toma valores de un espacio muestral; las probabilidades describen cuáles valores y conjunto de valores son más probables.
<p>&nbsp;</p>

- ***Evento, $E$:*** conjunto de valores posibles (resultados) de una variable aleatoria que ocurre con cierta probabilidad.
<p>&nbsp;</p>

- ***[Función de probabilidad](https://en.wikipedia.org/wiki/Probability_distribution) o medida de probabilidad:*** describe la probabilidad $P(X \in E)$ de que ocurra el evento $E$.
<p>&nbsp;</p>

- ***[Función de distribución acumulativa](https://en.wikipedia.org/wiki/Cumulative_distribution_function):*** función que evalúa la probabilidad de que $X$ tome un valor menor o igual a $x$ para una variable aleatoria (solo para variables aleatorias con valores reales).
<p>&nbsp;</p>

- ***Función cuantil:*** la inversa de la función de distribución acumulativa. Dado $x$ tal que, con probabilidad $q$, $X$ no superará $x$.
<p>&nbsp;</p>

- ***Grados de libertad:*** Los grados de libertad representan el número de valores que son libres de variar en el contexto de una estadística o de un conjunto de datos. En otras palabras, los grados de libertad indican cuánta variabilidad o flexibilidad hay en un sistema o en una muestra de datos. En términos más simples, los grados de libertad se refieren a la cantidad de información que se puede utilizar para hacer inferencias o estimaciones en un problema estadístico. A menudo, los grados de libertad se asocian con el tamaño de una muestra o con el número de restricciones impuestas en un análisis.

### Variables Continuas

- Las variables continuas pueden tomar un número infinito de valores dentro de un rango específico.
<p>&nbsp;</p>

- Los valores pueden ser números reales.
<p>&nbsp;</p>

- Ejemplos de distribuciones continuas incluyen la distribución normal, la distribución exponencial y la distribución uniforme continua.
<p>&nbsp;</p>

- Se utilizan para modelar fenómenos que pueden tomar una amplia gama de valores.

### Variables Discretas

- Las variables discretas solo pueden tomar valores específicos y contables, generalmente números enteros.
<p>&nbsp;</p>

- Ejemplos de distribuciones discretas incluyen la distribución binomial, la distribución de Poisson y la distribución de Bernoulli.
<p>&nbsp;</p>

- Se utilizan para modelar eventos que tienen resultados específicos y no continuos.

### Función de Probabilidad y Función de Distribución Acumulativa

#### Introducción

La [Función de probabilidad](https://en.wikipedia.org/wiki/Probability_distribution) describe la probabilidad de que una variable aleatoria discreta tome un valor específico. En otras palabras, asigna probabilidades a valores individuales en el rango de la variable aleatoria discreta. La función de probabilidad asigna probabilidades a valores discretos de la variable aleatoria, es decir, valores aislados en el espacio muestral. Se denota como $P(X=x)$, donde $X$ es la variable aleatoria y $x$ es un valor particular en su rango. Por ejemplo, si lanzamos un dado justo, la función de probabilidad para la variable aleatoria que representa el resultado del lanzamiento asignaría probabilidades iguales de $1/6$ a cada uno de los valores del $1$ al $6$.

En estadísticas, representamos la distribución de variables discretas con [Funciones de Masa de Probabilidad](https://en.wikipedia.org/wiki/Probability_mass_function) (*PMF*, por sus siglas en inglés) y [Funciones de Distribución Acumulativa](https://en.wikipedia.org/wiki/Cumulative_distribution_function) (*CDF*, por sus siglas en inglés). Representamos las distribuciones de variables continuas con [Funciones de Densidad de Probabilidad](https://en.wikipedia.org/wiki/Probability_density_function) (*PDF*, por sus siglas en inglés) y *CDF's*.

La *PMF* define la probabilidad de todos los posibles valores $x$ de la variable aleatoria. Una *PDF* es lo mismo, pero para valores continuos. El *CDF* representa la probabilidad de que la variable aleatoria $X$ tenga un resultado menor o igual al valor $x$. El nombre *CDF* se utiliza tanto para distribuciones discretas como continuas.

Las funciones que describen las *PMF*, *PDF* y *CDF* pueden ser bastante intimidantes al principio, pero sus contrapartes visuales a menudo parecen bastante intuitivas.

La [Función de probabilidad](https://en.wikipedia.org/wiki/Probability_distribution) y la [Función de distribución acumulativa](https://en.wikipedia.org/wiki/Cumulative_distribution_function) son dos conceptos importantes en estadísticas y teoría de probabilidad, y cada una tiene un propósito específico en la descripción y análisis de variables aleatorias. Aquí se explica la diferencia entre ambas:

#### Función de Masa de Probabilidad (PMF)

La [Función de Masa de Probabilidad](https://en.wikipedia.org/wiki/Probability_mass_function) (*PMF*, por sus siglas en inglés, *Probability Mass Function*) es una función estadística que se utiliza para describir la distribución de probabilidad de una variable discreta. En otras palabras, la *PMF* asigna una probabilidad a cada posible valor discreto que puede tomar una variable aleatoria. La *PMF* se utiliza comúnmente para representar y analizar variables aleatorias discretas, como el resultado de lanzar un dado, el número de éxitos en una serie de *ensayos de Bernoulli* o el número de defectos en un lote de productos.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# PMF Visualization
n = 100
p = 0.5

fig, ax = plt.subplots(1, 1, figsize=(17,5))
x = np.arange(binom.ppf(0.01, n, p), binom.ppf(0.99, n, p))
ax.plot(x, binom.pmf(x, n, p), 'bo', ms=8, label='Binomial PMF')
ax.vlines(x, 0, binom.pmf(x, n, p), colors='b', lw=5, alpha=0.5)
rv = binom(n, p)
#ax.vlines(x, 0, rv.pmf(x), colors='k', linestyles='-', lw=1, label='frozen PMF')
ax.legend(loc='best', frameon=False, fontsize='xx-large')
plt.title('PMF of a binomial distribution (n=100, p=0.5)', fontsize='xx-large')
plt.show()
```

#### Función de Densidad de Probabilidad (Probability Density Function, PDF):

A diferencia de las *Funciones de Masa de Probabilidad (PMF)* que se utilizan para variables discretas, las [Funciones de Densidad de Probabilidad](https://en.wikipedia.org/wiki/Probability_density_function) (*PDF*, por sus siglas en inglés, *Probability Density Functions*) se aplican a variables aleatorias continuas.

La *PDF* describe la distribución de probabilidad de una variable aleatoria continua y especifica cómo se distribuyen las densidades de probabilidad en el rango de valores posibles de la variable continua. A diferencia de las *PMF*, que asignan probabilidades directas a valores individuales, las *PDF* no asignan probabilidades a valores individuales, sino que describen cómo se agrupa la probabilidad en intervalos de valores.

Algunos aspectos clave de las PDF incluyen:

- ***Función de Densidad:*** La *PDF* se denota generalmente como $f(x)$ y se interpreta como la densidad de probabilidad en el punto $x$. No representa probabilidades directas, sino la densidad de probabilidad en un intervalo infinitesimal alrededor de $x$.
<p>&nbsp;</p>

- ***Integral Total:*** La integral de la *PDF* sobre todo su dominio debe ser igual a $1$, es decir, 

$$\int_{-\infty}^{\infty} f(x) dx = 1$$

Esto garantiza que la probabilidad total bajo la curva de la *PDF* sea igual a $1$.

- ***Valores No Negativos:*** La *PDF* debe ser no negativa para todos los valores $x$, es decir, $f(x) \geq 0$ para todo $x$ en su dominio.
<p>&nbsp;</p>

- ***Área bajo la Curva:*** El área bajo la curva de la *PDF* en un intervalo particular corresponde a la probabilidad de que la variable aleatoria caiga dentro de ese intervalo. Es decir, la probabilidad de que $X$ esté en el intervalo $[a, b]$ se calcula como $\int_{a}^{b} f(x) dx$.
<p>&nbsp;</p>

- ***Valores en el Eje Horizontal:*** A diferencia de las *PMF*, las *PDF* pueden tener valores en el eje horizontal (eje $x$) que no corresponden necesariamente a valores específicos de la variable aleatoria. En lugar de asignar probabilidades a puntos individuales, las *PDF* representan cómo se agrupan las probabilidades en intervalos continuos.
<p>&nbsp;</p>

- ***Representación Gráfica:*** Las *PDF* se representan gráficamente mediante curvas continuas. El área bajo la curva en un intervalo representa la probabilidad de que la variable aleatoria caiga dentro de ese intervalo.

Un ejemplo común de *PDF* es la de la *distribución normal* (también conocida como *distribución gaussiana*). La curva de la *PDF* de una *distribución normal* es una forma de campana y se utiliza ampliamente en estadísticas debido a su aplicabilidad en una variedad de contextos.

En resumen, las *Funciones de Densidad de Probabilidad* (*PDF*) son herramientas esenciales para describir y comprender la distribución de probabilidad de variables aleatorias continuas. Ayudan a calcular probabilidades en intervalos, entender tendencias y caracterizar el comportamiento de variables continuas en estadísticas y análisis de datos.

```python
import scipy.stats as stats
# Plot normal distribution
mu = 0
variance = 1
sigma = np.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.figure(figsize=(16,5))
plt.plot(x, stats.norm.pdf(x, mu, sigma), label='Normal Distribution')
plt.title('Normal Distribution with mean = 0 and std = 1')
plt.legend(fontsize='xx-large')
plt.show()
```

#### Función de Distribución Acumulativa (Cumulative Distribution Function - CDF)

La [Función de distribución acumulativa](https://en.wikipedia.org/wiki/Cumulative_distribution_function) (*CDF*, por sus siglas en inglés, *Cumulative Distribution Function*) es una función estadística que se utiliza tanto para variables discretas como continuas. La *CDF* proporciona información sobre la probabilidad acumulativa de que una variable aleatoria tome un valor igual o menor que un valor dado. Es una medida acumulativa de la probabilidad a medida que avanzamos a lo largo del rango de valores posibles de la variable.

Para una variable aleatoria discreta, la *CDF* se define como:

$$
F(x) = P(X \leq x)
$$

Donde:
- $F(x)$ es la probabilidad acumulativa de que la variable $X$ sea menor o igual a $x$.
<p>&nbsp;</p>

- $X$ es la variable aleatoria discreta.
<p>&nbsp;</p>

- $x$ es el valor hasta el cual se está calculando la probabilidad acumulativa.

Para una variable aleatoria continua, la *CDF* se define como:

$$
F(x) = \int_{-\infty}^{x} f(t) dt
$$

Donde:
- $F(x)$ es la probabilidad acumulativa de que la variable $X$ sea menor o igual a $x$.
<p>&nbsp;</p>

- $f(t)$ es la *Función de Densidad de Probabilidad* (*PDF*) de la variable $X$.
<p>&nbsp;</p>

- $X$ es la variable aleatoria continua.
<p>&nbsp;</p>

- $x$ es el valor hasta el cual se está calculando la probabilidad acumulativa.

Algunas características clave de la *CDF* incluyen:

1. ***Valores en el Eje Vertical:*** La *CDF* proporciona valores acumulativos de probabilidad en el eje vertical (eje $y$). En un punto particular $x$, $F(x)$ representa la probabilidad acumulativa de que la variable aleatoria sea menor o igual a $x$.
<p>&nbsp;</p>

2. ***Valores en el Eje Horizontal:*** El eje horizontal (eje $x$) contiene los valores posibles de la variable aleatoria. A medida que avanzamos a lo largo del eje $x$, la *CDF* acumula probabilidades.
<p>&nbsp;</p>

3. ***Límites de la CDF:*** La *CDF* tiene límites en $0$ y $1$. $F(-\infty) = 0$ y $F(\infty) = 1$, lo que significa que la probabilidad acumulativa en el extremo inferior del rango es $0$ y en el extremo superior es $1$.
<p>&nbsp;</p>

4. ***Monotonicidad:*** La *CDF* es una función monótona no decreciente, lo que significa que nunca disminuye a medida que avanzamos a lo largo del eje $x$.
<p>&nbsp;</p>

5. ***Saltos en Caso de Variables Discretas:*** En el caso de variables aleatorias discretas, la *CDF* puede tener saltos discretos en puntos donde la variable cambia de valor.

La *CDF* es una herramienta fundamental en estadísticas y teoría de la probabilidad. Se utiliza para calcular probabilidades acumulativas, determinar percentiles y comprender la distribución de probabilidad de una variable aleatoria en su totalidad. La representación gráfica de la *CDF* se realiza trazando una curva acumulativa que muestra cómo se acumulan las probabilidades a medida que avanzamos a lo largo del rango de valores de la variable.

```python
# Data
X  = np.arange(-2, 2, 0.01)
Y  = np.exp(-X ** 2)

# Normalize data
Y = Y / (0.01 * Y).sum()

# Plot the PDF and CDF
plt.figure(figsize=(15,5))
plt.title('Continuous Normal Distributions', fontsize='xx-large')
plt.plot(X, Y, label='Probability Density Function (PDF)')
plt.plot(X, np.cumsum(Y * 0.01), 'r', label='Cumulative Distribution Function (CDF)')
plt.legend(fontsize='xx-large')
plt.show()
```

## Distribuciones de Variables Continuas

### Distribución Normal (Gaussiana)

La [distribución normal](https://en.wikipedia.org/wiki/Normal_distribution), también conocida como distribución gaussiana, es una de las distribuciones más importantes en estadística. Se caracteriza por su forma de campana y es ampliamente utilizada debido al [Teorema Central del Límite](https://en.wikipedia.org/wiki/Central_limit_theorem), que establece que las sumas de variables aleatorias independientes tienden a seguir una distribución normal. Está completamente definida por dos parámetros: la media ($\mu$) y la desviación estándar ($\sigma$). Se denota como N($\mu$, $\sigma^2$), donde $\mu$ es la media y $\sigma^2$ es la varianza. 

En estadísticas, una distribución normal o distribución gaussiana es un tipo de distribución de probabilidad continua para una variable aleatoria de valores reales. La forma general de su función de densidad de probabilidad es:

$$P(x)=\frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{1}{2} \left(\frac{x-\mu}{\sigma} \right)^2}$$

Imagina que estamos midiendo la altura de muchas personas en una población. En este caso, es más probable encontrar personas que tengan una altura promedio, y a medida que nos alejamos de la altura promedio hacia arriba o hacia abajo, la probabilidad de encontrar a alguien disminuye gradualmente. Esto forma una distribución normal, que tiene una forma de campana.

***Características importantes:***

<p float="center">
  <img src="https://github.com/carlosalvarezh/Curso_CEC_EAFIT/blob/main/images/C02_05Standard_deviation_diagram_micro.svg?raw=true" width="350" />
</p>


- ***Forma Simétrica:*** La distribución normal es simétrica respecto a su media, $\mu$. Esto significa que la mitad de la curva se encuentra a la izquierda y la otra mitad a la derecha de la media, y ambas partes son idénticas en forma.
<p>&nbsp;</p>

- ***Media, Mediana y Moda Igual:*** En una distribución normal, la mediana y la moda coinciden en el valor de la media ($\mu$), lo que demuestra su simetría.
<p>&nbsp;</p>

- ***68-95-99.7 Regla:*** Aproximadamente el $68\%$ de los datos se encuentran dentro de una desviación estándar de la media, $[\mu-\sigma, \mu +\sigma]$. Alrededor del $95\%$ dentro de dos desviaciones estándar,$[\mu-2\sigma, \mu +2\sigma]$, y aproximadamente el $99.7\%$ dentro de tres desviaciones estándar, $[\mu-3\sigma, \mu +3\sigma]$.
<p>&nbsp;</p>

- ***Distribución Estándar:*** La [distribución normal estándar ($Z$)](https://en.wikipedia.org/wiki/Standard_normal_table) es una versión especial con media $0$ y desviación estándar $1$. Cualquier distribución normal puede ser transformada en la distribución normal estándar mediante la fórmula 

$$Z = \frac{(x - \mu)}{\sigma}$$

***Uso:***
- Modela datos que son aproximadamente simétricos y continúos.
<p>&nbsp;</p>

- Se utiliza en pruebas de hipótesis, intervalos de confianza y análisis de regresión.


***Limitaciones:***

- No es adecuada para datos muy sesgados o no normales.
<p>&nbsp;</p>

- No es apropiada para datos discretos.

veamos un código en python que grafica la función Normal con sus principales características

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parámetros de la distribución normal estandard
mu = 0     # Media
sigma = 1  # Desviación estándar

# Generar datos de la distribución normal
datos = np.random.normal(mu, sigma, 1000)

# Crear el histograma
plt.figure(figsize=(8, 4))
plt.hist(datos, bins=30, density=True, alpha=0.6, color='b', label='Histograma')

# Crear la curva de densidad de probabilidad teórica
x = np.linspace(-4.0, 4.0, 100)
pdf = stats.norm.pdf(x, mu, sigma)
plt.plot(x, pdf, 'r-', label='Distribución Normal')

# Línea vertical para la media
plt.axvline(mu, color='k', linestyle='dashed', linewidth=1, label='$\mu (media)$')

# Líneas verticales para ±1 desviación estándar
std_dev = sigma
plt.axvline(mu - std_dev, color='g', linestyle='dashed', linewidth=1, label='$\mu \pm \sigma$')
plt.axvline(mu + std_dev, color='g', linestyle='dashed', linewidth=1)
plt.axvline(mu - 2*std_dev, color='b', linestyle='dashed', linewidth=1, label='$\mu \pm 2\sigma$')
plt.axvline(mu + 2*std_dev, color='b', linestyle='dashed', linewidth=1)
plt.axvline(mu - 3*std_dev, color='r', linestyle='dashed', linewidth=1, label='$\mu \pm 3\sigma$')
plt.axvline(mu + 3*std_dev, color='r', linestyle='dashed', linewidth=1)

# Configurar el gráfico
plt.title('Distribución Normal')
plt.xlabel('Valores')
plt.ylabel('Densidad de Probabilidad')
plt.legend()

# Mostrar el gráfico
plt.grid(True)
plt.show()
```

### Distribución Exponencial

La [distribución exponencial](https://en.wikipedia.org/wiki/Exponential_distribution) es una distribución de probabilidad continua que se utiliza comúnmente para modelar el tiempo entre eventos en procesos de llegada aleatoria o procesos de decaimiento exponencial. Es especialmente útil cuando se trata de eventos que ocurren de manera independiente a una tasa constante en el tiempo o en el espacio. Por lo tanto, es ampliamente aplicable en áreas como la teoría de colas, la fiabilidad, la ingeniería, la economía y la biología, entre otras.

La función de densidad de probabilidad (FDP) de la distribución exponencial está definida de la siguiente manera para una variable aleatoria continua $X$:

$$
P(x|\lambda)= \left\{ \begin{array}{lcc} \lambda e^{-\lambda x} & si & x \geq 0 \\ \\ 0 & si & x < 0 \end{array} \right.
$$

Donde:
- $P(x∣\lambda)$ es la FDP de $X$.
<p>&nbsp;</p>

- $\lambda$ (lambda) es el parámetro de tasa, que es una medida de la frecuencia de ocurrencia de eventos. Debe ser un valor positivo ($\lambda>0$).
<p>&nbsp;</p>

- $e$ es la base del logaritmo natural ($\approx 2.71828$).

***Características:***

- ***Media:*** La media o valor esperado de una variable exponencial con parámetro $\lambda$ es igual a$1/\lambda$, es decir, $\mu = 1/\lambda$
<p>&nbsp;</p>

- ***Varianza:*** La varianza de una variable exponencial con parámetro $\lambda$ es igual a $1/\lambda^2$, es decir, $\sigma^2 = 1/ \lambda^2$.
<p>&nbsp;</p>

- ***Falta de Memoria:*** La distribución exponencial tiene una propiedad única llamada "falta de memoria". Esto significa que la probabilidad de que ocurra un evento en el futuro no depende de cuánto tiempo ha pasado sin que ocurra el evento. Matemáticamente, se expresa como $P(X>s+t∣X>s)=P(X>t)$ para $s,t>0$.

***Uso:***

La distribución exponencial se utiliza en una variedad de situaciones, incluyendo:

- Modelado de tiempos entre llegadas de clientes a un servicio.
<p>&nbsp;</p>

- Estudio de la vida útil de productos y componentes.
<p>&nbsp;</p>

- Análisis de confiabilidad y mantenimiento de equipos.
<p>&nbsp;</p>

- Modelado de tasas de fallos en sistemas.
<p>&nbsp;</p>

- Estudio de procesos de decaimiento radioactivo.
<p>&nbsp;</p>

- Análisis de tiempo hasta que ocurra un evento, como el tiempo hasta que un usuario haga clic en un anuncio en línea.


***Limitaciones:***

- La distribución exponencial asume que los eventos ocurren de manera independiente a una tasa constante. Esto puede no ser aplicable en situaciones donde la tasa de ocurrencia cambia con el tiempo.
<p>&nbsp;</p>

- No es adecuada para modelar eventos que tienen un componente de memoria, es decir, donde la probabilidad de que ocurra un evento puede depender de cuánto tiempo ha pasado desde la última ocurrencia.

***Ejemplo:***

A continuación, se presenta un ejemplo en Python que genera datos de una distribución exponencial y muestra una gráfica de su densidad de probabilidad utilizando las bibliotecas NumPy y Matplotlib:

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetro de tasa lambda
lambd = 0.5

# Generar datos de una distribución exponencial
data = np.random.exponential(scale=1/lambd, size=1000)

plt.figure(figsize=(8, 4))

# Histograma de los datos
plt.hist(data, bins=30, density=True, alpha=0.6, color='b', label='Datos')

# Función de densidad de probabilidad teórica
x = np.linspace(0, 10, 100)
pdf = lambd * np.exp(-lambd * x)
plt.plot(x, pdf, 'r-', lw=2, label='FDP')

plt.title('Distribución Exponencial')
plt.xlabel('Tiempo entre eventos')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

### Distribución Uniforme Continua

La [distribución uniforme continua](https://en.wikipedia.org/wiki/Continuous_uniform_distribution) es un tipo de distribución de probabilidad que se caracteriza por tener una densidad de probabilidad constante dentro de un rango específico de valores y cero fuera de ese rango. En otras palabras, esta distribución asigna igual probabilidad a todos los valores posibles dentro de un intervalo.

La FDP de la distribución uniforme continua se define de la siguiente manera:

$$
P(x|a,b)= \left\{ \begin{array}{lcc} \frac{1}{b-a} & \text{si } a \leq x \leq b \\ \\ 0 & \text{en otros casos} \end{array} \right.
$$

Donde:

- $P(x|a,b)$ es la FDP de la variable aleatoria $X$.
<p>&nbsp;</p>

- $a$ es el valor mínimo del intervalo.
<p>&nbsp;</p>

- $b$ es el valor máximo del intervalo.

***Características:***

- ***Media:*** La media de una distribución uniforme continua con valores mínimo $a$ y máximo $b$ se calcula como $\mu = \frac{a+b}{2}$
<p>&nbsp;</p>

- ***Varianza:*** La varianza de esta distribución se calcula como $\sigma^2 = \frac{(b-a)^2}{12}$.
<p>&nbsp;</p>

- ***Igual Probabilidad:*** Todos los valores dentro del intervalo $[a,b]$ tienen la misma probabilidad de ocurrencia, lo que significa que no hay valores más probables que otros en ese intervalo.

***Usos:***
La distribución uniforme continua se utiliza en diversas aplicaciones, incluyendo:

- Modelado de procesos aleatorios donde no existe sesgo hacia ningún valor en el intervalo.
<p>&nbsp;</p>

- Generación de números aleatorios en simulaciones y juegos.
<p>&nbsp;</p>

- Muestreo aleatorio simple, donde se seleccionan elementos de una población de manera equiprobable.
<p>&nbsp;</p>

- Cálculos de probabilidad en situaciones donde todos los resultados son igualmente probables.


***Limitaciones:***

- La distribución uniforme continua puede no ser adecuada para modelar situaciones en las que se espera que los valores sean más probables en ciertas partes del intervalo. En tales casos, otras distribuciones con forma más específica pueden ser más apropiadas.

***Ejemplo Aplicativo en Python:***

A continuación, se muestra un ejemplo en Python que genera datos de una distribución uniforme continua y crea un histograma para visualizar la uniformidad de los valores generados:

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el intervalo [a, b]
a = 0
b = 10

# Generar datos de una distribución uniforme continua
data = np.random.uniform(a, b, size=100000)

plt.figure(figsize=(8, 4))

# Histograma de los datos
plt.hist(data, bins=30, density=True, alpha=0.6, color='b', label='Datos')

# Función de densidad de probabilidad teórica
x = np.linspace(a, b, 100)
pdf = np.full_like(x, 1 / (b - a))  # Crear un array del mismo tamaño que x
plt.plot(x, pdf, 'r-', lw=2, label='FDP')

plt.title('Distribución Uniforme Continua')
plt.xlabel('Valores')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

### Distribución T de Student

La [distribución t de Student](https://en.wikipedia.org/wiki/Student%27s_t-distribution), también conocida simplemente como *distribución t*, es una *distribución de probabilidad* que se utiliza principalmente en estadísticas para realizar inferencias sobre las medias de poblaciones pequeñas o cuando la desviación estándar de la población es desconocida. Fue desarrollada por el estadístico *William Sealy Gosset* y se utiliza comúnmente en pruebas de hipótesis y análisis de intervalos de confianza.

La FDP de la *distribución t* con $n$ grados de libertad está definida de la siguiente manera:

$$
P(x|n)=\frac{\Gamma \left( \frac{n+1}{2} \right)}{\sqrt{n \pi} \Gamma \left( \frac{n}{2} \right)} \left( 1 + \frac{x^2}{n}\right)^{-\frac{n+1}{2}}
$$

Donde:

- $P(x|n)$ es la FDP de la variable aleatoria $t$
<p>&nbsp;</p>

- $n$ es el número de grados de libertad, que determina la forma de la distribución. Cuanto mayor sea $n$, más se asemejará la *distribución t* a una distribución normal estándar.
<p>&nbsp;</p>

- $\Gamma$ es la [función gamma](https://en.wikipedia.org/wiki/Gamma_function), que es una función matemática especial utilizada en el cálculo de integrales.

***Características:***

- ***Forma de Campana:*** Al igual que la *distribución normal*, la *distribución t* tiene una forma de campana, pero con colas más gruesas. Las colas más gruesas reflejan una mayor variabilidad debido a la incertidumbre sobre la desviación estándar de la población.
<p>&nbsp;</p>

- ***Grados de Libertad:*** El parámetro $n$ (grados de libertad) es fundamental en la *distribución t*. Cuantos más grados de libertad tenga, más se asemejará la *distribución t* a una *distribución normal estándar*. A medida que $n$ aumenta, las colas de la distribución se vuelven menos pesadas.
<p>&nbsp;</p>

- Las *medidas de tendencia central (Media, Moda, Mediana)* en una *distribución t de Student* se calculan de la misma manera que en otras distribuciones. Sin embargo, debido a las colas más gruesas de la *distribución t*, la media puede ser más sensible a valores atípicos en comparación con la mediana.

***Usos:***

La *distribución t* se utiliza en diversas aplicaciones, incluyendo:

- Pruebas de hipótesis para comparar medias de muestras pequeñas cuando la desviación estándar de la población es desconocida.
<p>&nbsp;</p>

- Cálculo de intervalos de confianza para estimar parámetros poblacionales, como la media, cuando se trabaja con muestras pequeñas.
<p>&nbsp;</p>

- Análisis de regresión, especialmente cuando se realiza regresión lineal y se desea evaluar la significancia de los coeficientes.
<p>&nbsp;</p>

- Evaluación de diferencias entre grupos en experimentos y estudios clínicos.

***Limitaciones:***

- La *distribución t* asume que los datos siguen una *distribución normal* en la población. Si los datos no cumplen con esta suposición, las inferencias basadas en la *distribución t* pueden ser incorrectas.
<p>&nbsp;</p>

- Es más sensible a valores atípicos (outliers) que la distribución normal estándar debido a sus colas más gruesas.


***Ejemplo Aplicativo en Python:***

A continuación, se presenta un ejemplo en Python que genera datos de una *distribución t* con 20 grados de libertad y crea un histograma para visualizar la forma de la *distribución t*:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Grados de libertad
df = 20

# Generar datos de una distribución t
data = np.random.standard_t(df, size=1000)

plt.figure(figsize=(8, 4))

# Histograma de los datos
plt.hist(data, bins=30, density=True, alpha=0.6, color='b', label='Datos')

# Función de densidad de probabilidad teórica
x = np.linspace(-4, 4, 1000)
pdf = t.pdf(x, df)
plt.plot(x, pdf, 'r-', lw=2, label='FDP')

plt.title('Distribución t de Student')
plt.xlabel('Valores')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

### Distribución Chi-Cuadrado

La [distribución chi-cuadrado](https://en.wikipedia.org/wiki/Chi-squared_distribution), denotada como $\chi^2$, es una distribución de probabilidad que se utiliza comúnmente en estadísticas para modelar la variabilidad de una población y realizar pruebas de hipótesis sobre la varianza o la dispersión de una muestra de datos. Esta distribución es especialmente útil cuando se trabaja con muestras pequeñas y se busca analizar la variabilidad en una población.

La FDP de la distribución chi-cuadrado con $k$ grados de libertad está definida de la siguiente manera:

$$
P(x|k)=\frac{1}{2^{\frac{x}{2}}\Gamma \left( \frac{k}{2} \right)}x^{\frac{k}{2}-1}e^{-\frac{x}{2}}
$$

Donde:

- $P(x|k)$ es la FDP de la variable aleatoria chi-cuadrado con $k$ grados de libertad.
<p>&nbsp;</p>

- $k$ es el número de grados de libertad, que determina la forma de la distribución. Cuanto mayor sea $k$, más se asemejará la distribución chi-cuadrado a una distribución normal.

***Características:***

- ***Forma de la Distribución:*** La *distribución chi-cuadrado* tiene una forma asimétrica y no negativa. Su forma se determina por el número de grados de libertad $k$. A medida que $k$ aumenta, la distribución se acerca cada vez más a una distribución normal.
<p>&nbsp;</p>

- ***Media y Varianza:*** La media de una distribución chi-cuadrado con $k$ grados de libertad es $k$, y la varianza es $2k$. Esto significa que tanto la media como la varianza dependen del número de grados de libertad.

***Usos:*** 

La distribución chi-cuadrado se utiliza en diversas aplicaciones, incluyendo:

- Pruebas de hipótesis sobre la varianza o la dispersión de una población. Por ejemplo, se utiliza en la prueba chi-cuadrado para evaluar si la varianza de una población es igual a un valor especificado.
<p>&nbsp;</p>

- Análisis de regresión, especialmente en la prueba de significancia de los coeficientes de regresión.
<p>&nbsp;</p>

- En estadísticas inferenciales, se usa en la construcción de intervalos de confianza para la varianza de una población.

***Limitaciones:***

- La distribución chi-cuadrado asume que las observaciones en la muestra son independientes y provienen de una población normalmente distribuida. Si esta suposición no se cumple, los resultados basados en la distribución chi-cuadrado pueden ser inadecuados.
<p>&nbsp;</p>

- La distribución chi-cuadrado es adecuada principalmente para datos continuos y no discretos.

***Ejemplo Aplicativo en Python:***

A continuación, se presenta un ejemplo en Python que genera datos de una distribución chi-cuadrado con 5 grados de libertad y crea un histograma para visualizar la forma de la distribución chi-cuadrado:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Grados de libertad
k = 5

# Generar datos de una distribución chi-cuadrado
data = np.random.chisquare(k, size=1000)

plt.figure(figsize=(8, 4))

# Histograma de los datos
plt.hist(data, bins=30, density=True, alpha=0.6, color='b', label='Datos')

# Función de densidad de probabilidad teórica
x = np.linspace(0, 20, 1000)
pdf = chi2.pdf(x, k)
plt.plot(x, pdf, 'r-', lw=2, label='FDP')

plt.title('Distribución Chi-cuadrado')
plt.xlabel('Valores')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

### Distribución Log-Normal

La [distribución log-normal](https://en.wikipedia.org/wiki/Log-normal_distribution) es una distribución de probabilidad continua que se utiliza para modelar variables aleatorias que tienen una relación logarítmica y donde los logaritmos de las variables siguen una distribución normal. Es especialmente útil cuando se trabaja con datos que tienen asimetría positiva (sesgados hacia la derecha) en su distribución.

La FDP de la distribución log-normal está definida de la siguiente manera:

$$
P(x|\mu, \sigma)=\frac{1}{x\sigma \sqrt{2 \pi}}e^{-\frac{\left( ln(x)-\mu \right)^2}{2\sigma^2}}
$$

Donde:
- $P(x|\mu, \sigma)$ es la FDP de la variable aleatoria log-normal.
<p>&nbsp;</p>

- $\mu$ es el parámetro de ubicación y representa la media del logaritmo de la variable.
<p>&nbsp;</p>

- $\sigma$ es el parámetro de escala y representa la desviación estándar del logaritmo de la variable.

***Características:***

- ***Asimetría Positiva:*** La distribución log-normal es conocida por su asimetría positiva, lo que significa que la mayoría de los valores se concentran en el lado izquierdo de la distribución y algunos valores extremadamente grandes pueden influir en la cola derecha.
<p>&nbsp;</p>

- ***Transformación Logarítmica:*** Si se toma el logaritmo natural (ln) de una variable log-normal, el resultado sigue una distribución normal. Esto hace que la distribución log-normal sea útil en situaciones donde las variables se multiplican juntas, ya que la suma de los logaritmos se comporta de manera más aditiva.

***Usos:***

La distribución log-normal se utiliza en diversas aplicaciones, incluyendo:

- Modelado de fenómenos en los que las variables tienen una relación multiplicativa.
<p>&nbsp;</p>

- Modelado de la distribución de ingresos, precios de activos financieros y otros datos económicos.
<p>&nbsp;</p>

- Estudio de tamaños de partículas en aerosoles y materiales dispersos.
<p>&nbsp;</p>

- En biología, para modelar el crecimiento de poblaciones y tamaños de organismos. También, la presión arterial y las tasas metabólicas en diversos organismos a menudo siguen una distribución log-normal.

***Limitaciones:***

- La distribución log-normal puede no ser apropiada si los datos tienen una relación aditiva en lugar de multiplicativa. En tales casos, una distribución normal puede ser más adecuada.
<p>&nbsp;</p>

- Al igual que con otras distribuciones, es importante verificar si los datos se ajustan adecuadamente a una distribución log-normal antes de aplicarla en un análisis.


***Ejemplo Aplicativo en Python:***

A continuación, se presenta un ejemplo en Python que genera datos de una distribución log-normal y crea un histograma para visualizar la forma de la distribución log-normal:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import lognorm

# Parámetros de las distribuciones log-normal
parametros = [(0, 1), (0, 0.25), (0, 0.5)]  # Lista de tuplas (mu, sigma)

# Valores para el eje x
x = np.linspace(0, 5, 1000)

# Crear una figura
plt.figure(figsize=(8, 4))

# Generar y graficar cada distribución log-normal en la misma gráfica
for mu, sigma in parametros:
    pdf = lognorm.pdf(x, s=sigma, scale=np.exp(mu))
    plt.plot(x, pdf, label=f'μ={mu}, σ={sigma}')

# Configuración de la gráfica
plt.title('Distribuciones Log-Normal')
plt.xlabel('Valores')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)

# Colocar etiquetas cerca de cada curva
for mu, sigma in parametros:
    max_pdf = np.max(lognorm.pdf(x, s=sigma, scale=np.exp(mu)))
    plt.text(mu + 0.05, max_pdf, f'μ={mu}, σ={sigma}', fontsize=12, verticalalignment='bottom')

plt.show()
```

### Distribución Binomial

La [Distribución Binomial](https://en.wikipedia.org/wiki/Binomial_distribution) es utilizada para modelar el número de éxitos (eventos de interés) en una serie de ensayos independientes, donde cada ensayo tiene dos resultados posibles (éxito o fracaso). Sus parámetros son el número de ensayos, $n$, y la probabilidad de éxito, $p$.

La FDP de la distribución Binomial se representa de la siguiente manera:

$$P(X=k) = \binom{n}{k} \cdot p^k \cdot (1-p)^{n-k}$$

Donde:

- $P(X = k)$ es la probabilidad de obtener exactamente $k$ éxitos en $n$ ensayos.
- $n$ es el número total de ensayos.
- $k$ es el número de éxitos que se busca.
- $p$ es la probabilidad de éxito en un solo ensayo.
- $1 - p$ es la probabilidad de fracaso en un solo ensayo.
- $\binom{n}{k}$ es el coeficiente binomial, que se calcula como $\binom{n}{k} = \frac{n!}{k!(n-k)!}$, donde $n!$ es el factorial de $n$.

***Características:***
- ***Variables Discretas:*** La variable aleatoria en la distribución binomial es discreta, lo que significa que solo puede tomar valores enteros.
<p>&nbsp;</p>

- ***Dos Resultados Posibles:*** Cada ensayo tiene dos resultados posibles: éxito o fracaso.
<p>&nbsp;</p>

- ***Independencia:*** Los ensayos deben ser independientes entre sí, lo que significa que el resultado de un ensayo no afecta el resultado de los demás.
<p>&nbsp;</p>

- ***Probabilidad de Éxito Constante:*** La probabilidad de éxito ($p$) debe ser constante en todos los ensayos.

***Usos:***

La distribución binomial se utiliza en una variedad de aplicaciones del mundo real donde estamos interesados en contar eventos de éxito/fracaso, y cada intento es independiente y tiene la misma probabilidad de éxito. Algunos ejemplos incluyen:

- ***Control de Calidad:*** Para determinar la probabilidad de que un lote de productos cumpla con ciertos estándares de calidad.
<p>&nbsp;</p>

- ***Biología:*** Para modelar eventos de genética, como la probabilidad de que cierto alelo se herede.
<p>&nbsp;</p>

- ***Finanzas:*** En el cálculo de probabilidades en operaciones financieras, como la probabilidad de que un inversor tenga ganancias en una serie de operaciones.
<p>&nbsp;</p>

- ***Estadísticas Deportivas:*** Para evaluar el desempeño de equipos o jugadores en deportes, como la probabilidad de que un jugador de béisbol conecte un jonrón en un juego.
<p>&nbsp;</p>

- ***Ventas:*** Determinar la probabilidad de que un número específico de clientes realice una compra en un sitio web durante un período determinado.

***Limitaciones:***

A pesar de su versatilidad, la distribución binomial tiene algunas limitaciones importantes:

- ***Requisito de Independencia:*** Los ensayos deben ser independientes entre sí, lo que no siempre es el caso en situaciones del mundo real.
<p>&nbsp;</p>

- ***Requisito de Probabilidad Constante:*** La probabilidad de éxito ($p$) debe ser constante en todos los ensayos, lo que puede no ser realista en algunas situaciones.
<p>&nbsp;</p>

- ***Valores Discretos:*** Solo puede utilizarse para variables aleatorias discretas con dos resultados posibles.

***Ejemplo:***

Supongamos que lanzamos una moneda justa 10 veces y queremos calcular la probabilidad de obtener exactamente 3 caras. Aquí está el código Python para hacerlo utilizando la distribución binomial:

***Solución analítica:*** La fórmula general de probabilidad para la distribución binomial es:

$$P(X = k) = \binom{n}{k} \cdot p^k \cdot (1 - p)^{n - k}$$

Donde:
- $P(X = k)$ es la probabilidad de obtener $k$ éxitos en $n$ ensayos.
<p>&nbsp;</p>

- $\binom{n}{k}$ es el coeficiente binomial, que representa el número de formas de elegir $k$ éxitos de $n$ ensayos y se calcula como $\binom{n}{k} = \frac{n!}{k! (n - k)!}$.
<p>&nbsp;</p>

- $p$ es la probabilidad de éxito en un ensayo individual.
<p>&nbsp;</p>

- $(1 - p)$ es la probabilidad de fracaso en un ensayo individual.

Pasos de solución:

1. Calcular el coeficiente binomial $\binom{n}{k}$:
   $$\binom{10}{3} = \frac{10!}{3! \cdot (10 - 3)!} = \frac{10!}{3! \cdot 7!} = \frac{10 \cdot 9 \cdot 8}{3 \cdot 2 \cdot 1} = 120$$
<p>&nbsp;</p>

2. Aplicar la fórmula de probabilidad de la distribución binomial:

$$P(X = 3) = 120 \cdot (0.5)^3 \cdot (1 - 0.5)^{10 - 3} = 120 \cdot 0.125 \cdot 0.125 = 0.1172 = 11.7\%$$

Por lo tanto, la probabilidad de obtener exactamente $3$ caras en $10$ lanzamientos de una moneda justa es del $11.7%$.

Este es un ejemplo simple de cómo resolver un problema de distribución binomial de forma analítica utilizando la fórmula de probabilidad binomial. 

***Solución en Python:***

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

n = 10  # Número de ensayos (lanzamientos de la moneda)
p = 0.5  # Probabilidad de éxito (cara en una moneda justa)
k = 3  # Número de éxitos deseados (caras)

# Cálculo de la probabilidad usando la distribución binomial
probability = binom.pmf(k, n, p)

print(f'La probabilidad de obtener exactamente 3 caras en 10 lanzamientos es: {probability:.4f}')

# Gráfico de la distribución binomial
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)

# Crear una figura
plt.figure(figsize=(8, 4))
plt.bar(x, pmf)
plt.title('Distribución Binomial')
plt.xlabel('Número de Caras')
plt.ylabel('Probabilidad')
plt.show()
```

### Distribución de Poisson

La [distribución de Poisson](https://en.wikipedia.org/wiki/Poisson_distribution) es una distribución de probabilidad discreta que modela la probabilidad de que ocurra un número específico de eventos raros en un intervalo de tiempo o espacio dado. Esta distribución es especialmente útil cuando se trata de contar eventos que ocurren de manera independiente y a una tasa constante promedio.

La ecuación de la distribución de Poisson se expresa de la siguiente manera:

$$P(X = k) = \frac{e^{-\lambda} \cdot \lambda^k}{k!}$$

Donde:
- $P(X = k)$ es la probabilidad de que ocurran $k$ eventos en el intervalo.
- $\lambda$ es la tasa promedio de eventos en el intervalo.
- $e$ es el [número de Euler](https://en.wikipedia.org/wiki/E_(mathematical_constant)), constante matemática aproximadamente igual a $2.71828$.
- $k$ es el número de eventos que se busca.
- $k!$ es el factorial de $k$.

***Características***

- **Variables Discretas**: La variable aleatoria en la distribución de Poisson es discreta, lo que significa que solo puede tomar valores enteros no negativos.
<p>&nbsp;</p>

- **Eventos Raros**: Se utiliza para modelar eventos raros que ocurren de manera independiente.
<p>&nbsp;</p>

- **Tasa Promedio Constante**: La tasa promedio ($\lambda$) de eventos debe ser constante en todo el intervalo.

***Usos***

La distribución de Poisson se aplica en diversas áreas, incluyendo:

- ***Procesos de Conteo***: Modelar eventos como llamadas telefónicas a un centro de atención al cliente, accidentes de tráfico en una intersección, errores de impresión en un libro, etc.
<p>&nbsp;</p>

- ***Biología***: Estudiar el número de mutaciones en una cadena de ADN, la cantidad de células cancerosas en un tejido, etc.
<p>&nbsp;</p>

- ***Ingeniería***: Analizar la ocurrencia de fallos en un sistema, como la cantidad de bombillas que se queman en un periodo de tiempo.
<p>&nbsp;</p>

- ***Física de Partículas***: Modelar el número de partículas radiactivas desintegrándose en un cierto tiempo.

***Limitaciones:***

A pesar de su utilidad, la distribución de Poisson también tiene limitaciones:

- ***Eventos Independientes***: Requiere que los eventos sean independientes entre sí, lo que puede no ser el caso en algunas situaciones.
<p>&nbsp;</p>

- ***Tasa Promedio Constante***: La tasa promedio ($\lambda$) debe ser constante en todo el intervalo, lo que puede no ser realista en algunas aplicaciones.

***Ejemplo Aplicativo en Python:***

Supongamos que en una intersección de carretera se produce un promedio de 2 accidentes por semana. Queremos calcular la probabilidad de que ocurran exactamente 3 accidentes en una semana utilizando la distribución de Poisson en Python:

```python
import matplotlib.pyplot as plt
from scipy.stats import poisson

lmbda = 2  # Tasa promedio de accidentes por semana
k = 3  # Número de accidentes que se busca

# Calcula la probabilidad
probabilidad = poisson.pmf(k, lmbda)

# Gráfica de la distribución de Poisson
x = range(0, 10)
y = [poisson.pmf(i, lmbda) for i in x]

# Crear una figura
plt.figure(figsize=(8, 4))
plt.bar(x, y)
plt.xlabel('Número de Accidentes')
plt.ylabel('Probabilidad')
plt.title('Distribución de Poisson')
plt.show()

print(f'La probabilidad de que ocurran exactamente {k} accidentes en una semana es {probabilidad:.4f}')
```

### Distribución de Bernoulli

La [distribución de Bernoulli](https://en.wikipedia.org/wiki/Bernoulli_distribution) es una distribución de probabilidad discreta que modela un experimento con dos resultados posibles: éxito (éxito) y fracaso (falla). Es una distribución de probabilidad simple pero fundamental, donde se describe la probabilidad de éxito en un solo ensayo independiente.

La ecuación de la distribución de Bernoulli se expresa de la siguiente manera:

$$P(X = x) = 
\begin{cases} 
p & \text{si } x = 1 \text{ (éxito)} \\
1 - p & \text{si } x = 0 \text{ (fracaso)}
\end{cases}
$$

Donde:
- $P(X = x)$ es la probabilidad de obtener el resultado $x$ (éxito o fracaso).
- $x$ es el resultado, que puede ser 1 (éxito) o 0 (fracaso).
- $p$ es la probabilidad de éxito en un solo ensayo.

***Características***

- ***Variable Discreta:*** La variable aleatoria en la distribución de Bernoulli es discreta y solo puede tomar dos valores: 0 o 1.
<p>&nbsp;</p>

- ***Experimento de Dos Resultados:*** Modela un experimento con dos resultados posibles: éxito o fracaso.
<p>&nbsp;</p>

- ***Probabilidad Constante:*** La probabilidad de éxito ($p$) es constante en todos los ensayos.

***Usos***

La distribución de Bernoulli se utiliza en diversas aplicaciones, como:

- ***Modelado de Eventos Binarios:*** Modela eventos que tienen solo dos resultados posibles, como el lanzamiento de una moneda (cara o cruz), el resultado de una prueba (aprobado o reprobado), etc.
<p>&nbsp;</p>

- ***Procesos de Decisión:*** En teoría de juegos y toma de decisiones, se utiliza para representar decisiones binarias.
<p>&nbsp;</p>

- ***Control de Calidad:*** Para determinar si un producto cumple con un estándar de calidad o no.

***Limitaciones***

Aunque la distribución de Bernoulli es útil para modelar eventos binarios, tiene limitaciones importantes:

- ***Solo Dos Resultados Posibles:*** Solo es adecuada para eventos con dos resultados posibles (éxito o fracaso).
<p>&nbsp;</p>

- ***Probabilidad Constante:*** Requiere que la probabilidad de éxito ($p$) sea constante en todos los ensayos, lo que puede no ser realista en algunas situaciones.

***Ejemplo Aplicativo en Python***

Supongamos que estamos modelando el lanzamiento de una moneda no sesgada, donde cara se considera un éxito (éxito) y cruz se considera un fracaso (falla). Queremos calcular la probabilidad de obtener una cara en un solo lanzamiento y generar una gráfica de la distribución de Bernoulli en Python:



```python
import matplotlib.pyplot as plt

p = 0.5  # Probabilidad de obtener una cara (éxito)

# Probabilidad de éxito (cara)
probabilidad_exito = p

# Probabilidad de fracaso (cruz)
probabilidad_fracaso = 1 - p

# Crear una figura
plt.figure(figsize=(8, 4))

# Gráfica de la distribución de Bernoulli
resultados = [0, 1]
probabilidades = [probabilidad_fracaso, probabilidad_exito]

plt.bar(resultados, probabilidades)
plt.xlabel('Resultado')
plt.ylabel('Probabilidad')
plt.xticks(resultados, ['Fracaso', 'Éxito'])
plt.title('Distribución de Bernoulli')
plt.show()

print(f'La probabilidad de obtener una cara en un solo lanzamiento es {probabilidad_exito:.2f}')
```

### Distribución Geométrica

La [distribución geométrica](https://en.wikipedia.org/wiki/Geometric_distribution) es una distribución de probabilidad discreta que modela el número de ensayos independientes necesarios para obtener el primer éxito en una secuencia de ensayos de Bernoulli. En otras palabras, se utiliza para calcular la probabilidad de que ocurra el primer éxito en el ensayo repetido de un evento binario hasta que se obtenga un éxito.

***Ecuación de la Distribución Geométrica***

La ecuación de la distribución geométrica se expresa de la siguiente manera:

$$P(X = k) = (1 - p)^{k-1} \cdot p$$

Donde:
- $P(X = k)$ es la probabilidad de obtener el primer éxito en el ensayo número $k$.
- $k$ es el número de ensayos necesarios para obtener el primer éxito.
- $p$ es la probabilidad de éxito en un solo ensayo de Bernoulli.

***Características***

- ***Variable Discreta:*** La variable aleatoria en la distribución geométrica es discreta y toma valores enteros no negativos.
<p>&nbsp;</p>

- ***Secuencia de Éxitos:*** Modela la secuencia de ensayos de Bernoulli hasta que se obtiene el primer éxito.
<p>&nbsp;</p>

- ***Probabilidad Constante:*** La probabilidad de éxito ($p$) en un solo ensayo de Bernoulli es constante.

***Usos***

La distribución geométrica se utiliza en diversas aplicaciones, que incluyen:

- ***Análisis de Procesos de Espera:*** Modela el tiempo que se tarda en recibir el primer cliente en un servidor, el primer accidente en una intersección, etc.
<p>&nbsp;</p>

- ***Estudio de Juegos de Azar:*** Se aplica en juegos de azar, como el póker, donde se puede utilizar para calcular la probabilidad de obtener una mano ganadora en el primer intento.
<p>&nbsp;</p>

- ***Control de Calidad:*** Para determinar cuántos productos deben ser inspeccionados antes de encontrar uno defectuoso.

***Limitaciones***

A pesar de su utilidad, la distribución geométrica tiene algunas limitaciones:

- ***Independencia de Ensayos:*** Requiere que los ensayos de Bernoulli sean independientes entre sí.
<p>&nbsp;</p>

- ***Probabilidad Constante:*** La probabilidad de éxito ($p$) debe ser constante en todos los ensayos, lo que puede no ser realista en algunas situaciones.

***Ejemplo Aplicativo en Python***

Supongamos que estamos jugando a un juego en el que se lanza una moneda no sesgada hasta que obtenemos el primer "cara" como éxito. Queremos calcular la probabilidad de que se necesiten exactamente 3 lanzamientos para obtener el primer éxito y generar una gráfica de la distribución geométrica en Python:

```python
import matplotlib.pyplot as plt

p = 0.5  # Probabilidad de obtener una cara (éxito)

k = 3  # Número de lanzamientos para obtener el primer éxito

# Calcula la probabilidad
probabilidad = (1 - p) ** (k - 1) * p

# Gráfica de la distribución geométrica
x = range(1, 11)
y = [(1 - p) ** (i - 1) * p for i in x]

# Crear una figura
plt.figure(figsize=(8, 4))
plt.bar(x, y)
plt.xlabel('Número de Lanzamientos para el Primer Éxito')
plt.ylabel('Probabilidad')
plt.title('Distribución Geométrica')
plt.show()

print(f'La probabilidad de obtener el primer éxito en el lanzamiento número {k} es {probabilidad:.4f}')
```

### Distribución Hipergeométrica

La [distribución hipergeométrica](https://en.wikipedia.org/wiki/Hypergeometric_distribution) es una distribución de probabilidad discreta que se utiliza para modelar experimentos en los que se extraen objetos de un conjunto finito sin reemplazo. En esta distribución, se calcula la probabilidad de obtener un número específico de elementos de un tipo particular en una muestra sin reemplazo de un conjunto de elementos que contiene objetos de diferentes tipos.

La ecuación de la distribución hipergeométrica se expresa de la siguiente manera:

$$P(X = k) = \frac{\binom{N_1}{k} \cdot \binom{N_2 - N_1}{n - k}}{\binom{N_2}{n}}$$

Donde:
- $P(X = k)$ es la probabilidad de obtener exactamente $k$ elementos del tipo de interés en la muestra.
- $k$ es el número de elementos del tipo de interés que se busca en la muestra.
- $n$ es el tamaño de la muestra.
- $N_1$ es el número total de elementos del tipo de interés en el conjunto.
- $N_2$ es el tamaño total del conjunto.
- $\binom{N}{k}$ denota el coeficiente binomial, que se calcula como $\binom{N}{k} = \frac{N!}{k!(N-k)!}$.

***Características Principales***

- ***Variable Discreta:*** La variable aleatoria en la distribución hipergeométrica es discreta y toma valores enteros no negativos.
<p>&nbsp;</p>

- ***Muestreo sin Reemplazo:*** Se utiliza para modelar experimentos donde los elementos se extraen del conjunto sin reemplazo, lo que significa que no se pueden seleccionar elementos duplicados.
<p>&nbsp;</p>

- ***Dependencia en el Tamaño de la Muestra:*** La probabilidad de éxito depende del tamaño de la muestra y del número total de elementos del tipo de interés.

***Usos***

La distribución hipergeométrica se aplica en diversas áreas, que incluyen:

- ***Calidad y Control de Inventarios:*** Para calcular la probabilidad de seleccionar una cierta cantidad de productos defectuosos en una inspección de calidad sin reemplazo.
<p>&nbsp;</p>

- ***Biología:*** Modelar la probabilidad de seleccionar un número específico de organismos de una especie en una muestra sin reemplazo.
<p>&nbsp;</p>

- ***Estadísticas de Encuestas:*** Calcular la probabilidad de obtener un número particular de respuestas afirmativas o negativas en una encuesta de opinión.

***Limitaciones***

A pesar de su utilidad, la distribución hipergeométrica tiene algunas limitaciones:

- ***Muestreo sin Reemplazo:*** Requiere que los elementos se extraigan del conjunto sin reemplazo, lo que puede no ser aplicable en todas las situaciones.
<p>&nbsp;</p>

- ***Dependencia del Tamaño de la Muestra:*** La probabilidad de éxito varía según el tamaño de la muestra, lo que puede hacer que los cálculos sean complicados en algunos casos.

***Ejemplo Aplicativo en Python***

Supongamos que tenemos una urna con 20 bolas, de las cuales 8 son rojas y 12 son verdes. Queremos calcular la probabilidad de obtener exactamente 3 bolas rojas al seleccionar al azar 5 bolas sin reemplazo y generar una gráfica de la distribución hipergeométrica en Python:


```python
import matplotlib.pyplot as plt
import scipy.special

N1 = 8  # Número de bolas rojas en la urna
N2 = 20  # Número total de bolas en la urna
n = 5  # Tamaño de la muestra
k = 3  # Número de bolas rojas buscadas en la muestra

# Calcula la probabilidad
probabilidad = (scipy.special.comb(N1, k) * scipy.special.comb(N2 - N1, n - k)) / scipy.special.comb(N2, n)

# Gráfica de la distribución hipergeométrica
x = range(0, min(N1, n) + 1)
y = [(scipy.special.comb(N1, i) * scipy.special.comb(N2 - N1, n - i)) / scipy.special.comb(N2, n) for i in x]

# Crear una figura
plt.figure(figsize=(8, 4))
plt.bar(x, y)
plt.xlabel('Número de Bolas Rojas en la Muestra')
plt.ylabel('Probabilidad')
plt.title('Distribución Hipergeométrica')
plt.show()

print(f'La probabilidad de obtener exactamente {k} bolas rojas al seleccionar 5 bolas es {probabilidad:.4f}')
```

### Distribución de Multinomial

La [distribución multinomial](https://en.wikipedia.org/wiki/Multinomial_distribution) es una distribución de probabilidad generalización de la distribución binomial y se utiliza para modelar experimentos en los que se clasifican objetos en más de dos categorías mutuamente excluyentes. Mientras que la distribución binomial se enfoca en el número de éxitos y fracasos en dos categorías, la distribución multinomial se aplica cuando hay más de dos resultados posibles y se calcula la probabilidad de observar una secuencia específica de categorías.

La ecuación de la distribución multinomial se expresa de la siguiente manera:

$$P(X_1 = x_1, X_2 = x_2, \ldots, X_k = x_k) = \frac{n!}{x_1! \cdot x_2! \cdot \ldots \cdot x_k!} \cdot p_1^{x_1} \cdot p_2^{x_2} \cdot \ldots \cdot p_k^{x_k}$$

Donde:
- $P(X_1 = x_1, X_2 = x_2, \ldots, X_k = x_k)$ es la probabilidad de obtener $x_1$ observaciones en la categoría 1, $x_2$ observaciones en la categoría 2, y así sucesivamente.
- $n$ es el número total de observaciones o ensayos.
- $x_1, x_2, \ldots, x_k$ son el número de observaciones en cada una de las $k$ categorías.
- $p_1, p_2, \ldots, p_k$ son las probabilidades de ocurrencia de cada categoría.
- $x_1!$ es el factorial de $x_1$, y lo mismo para las demás categorías.

***Características***

- ***Variable Discreta:*** La variable aleatoria en la distribución multinomial es discreta y toma valores enteros no negativos en cada categoría.
<p>&nbsp;</p>

- ***Múltiples Categorías*** Modela experimentos con más de dos categorías mutuamente excluyentes.
<p>&nbsp;</p>

- ***Probabilidades de Categoría Constantes:*** Las probabilidades de cada categoría ($p_1, p_2, \ldots, p_k$) deben ser constantes en todos los ensayos.

***Usos***

La distribución multinomial se utiliza en diversas aplicaciones, que incluyen:

- ***Estadísticas de Muestreo:*** Para modelar la distribución de categorías en una muestra de población, como la distribución de votos en una elección.
<p>&nbsp;</p>

- ***Biología:*** Para modelar la distribución de genotipos en una población de organismos, donde los genotipos son las categorías.
<p>&nbsp;</p>

- ***NLP (Procesamiento de Lenguaje Natural):*** En tareas de clasificación de texto, donde se clasifican documentos en múltiples categorías basadas en sus características.

***Limitaciones***

A pesar de su versatilidad, la distribución multinomial tiene algunas limitaciones:

- ***Independencia de Ensayos:*** Requiere que los ensayos sean independientes entre sí.
<p>&nbsp;</p>

- ***Probabilidades Constantes:*** Las probabilidades de categoría deben ser constantes en todos los ensayos.

***Ejemplo Aplicativo en Python***

Supongamos que estamos realizando una encuesta de satisfacción de clientes en una tienda de ropa, y tenemos tres categorías de respuestas posibles para cada pregunta: "Satisfecho", "Neutral" y "Insatisfecho". Queremos calcular la probabilidad de obtener 2 respuestas "Satisfecho", 3 respuestas "Neutral" y 1 respuesta "Insatisfecho" en una muestra de 6 respuestas y generar una gráfica de la distribución multinomial en Python:


```python
import matplotlib.pyplot as plt
from scipy.stats import multinomial

n = 6  # Tamaño de la muestra
p = [0.4, 0.3, 0.3]  # Probabilidades de cada categoría: Satisfecho, Neutral, Insatisfecho
x = [2, 3, 1]  # Número de respuestas en cada categoría

# Calcula la probabilidad
probabilidad = multinomial.pmf(x, n, p)

# Gráfica de la distribución multinomial
categorias = ['Satisfecho', 'Neutral', 'Insatisfecho']

# Crear una figura
plt.figure(figsize=(8, 4))
plt.bar(categorias, probabilidad)
plt.xlabel('Categorías')
plt.ylabel('Probabilidad')
plt.title('Distribución Multinomial')
plt.show()

print(f'La probabilidad de obtener 2 respuestas "Satisfecho", 3 respuestas "Neutral" y 1 respuesta "Insatisfecho" en la muestra es {probabilidad:.4f}')
```
